```python
import matplotlib.pyplot as plt
from PIL import Image
import base64

def analyze_lsb_blocks(image_path, channel="R", block_size=50, threshold=0.60):
    """
    Анализ распределения LSB в блоках изображения.
    :param image_path: Путь к изображению.
    :param channel: Цветовой канал для извлечения ("R", "G", "B").
    :param block_size: Размер блока .
    :param threshold: Пороговое значение доли битов 1 в LSB для выделения.
    :return: Извлеченный поток бит.
    """
    try:
        img = Image.open(image_path)
        pixels = img.load()
    except Exception as e:
        print(f"Не удалось открыть файл изображения: {e}")
        return []

    width, height = img.size
    num_blocks_x = width // block_size
    num_blocks_y = height // block_size

    lsb_ratios = []
    lsb_data = []

    for block_y in range(num_blocks_y):
        for block_x in range(num_blocks_x):
            lsb_zeros = 0
            lsb_ones = 0
            block_bits = []
            for y in range(block_y * block_size, (block_y + 1) * block_size):
                for x in range(block_x * block_size, (block_x + 1) * block_size):
                    r, g, b = pixels[x, y]
                    if channel == "R":
                        lsb = r & 1
                    elif channel == "G":
                        lsb = g & 1
                    elif channel == "B":
                        lsb = b & 1
                    block_bits.append(lsb)
                    if lsb == 0:
                        lsb_zeros += 1
                    else:
                        lsb_ones += 1
            ratio = lsb_ones / (lsb_zeros + lsb_ones)
            lsb_ratios.append(ratio)
            if ratio > threshold:  # Извлекаем только выделенные блоки
                lsb_data.extend(block_bits)

    # Визуализация тепловой карты
    lsb_ratios_2d = [
        lsb_ratios[i * num_blocks_x:(i + 1) * num_blocks_x]
        for i in range(num_blocks_y)
    ]
    plt.imshow([[r for r in row] for row in lsb_ratios_2d], cmap='hot', interpolation='nearest')
    plt.colorbar(label='Доля бит 1 в LSB')
    plt.title(f"Тепловая карта распределения LSB ({channel} канал)")
    plt.show()

    return lsb_data

def try_base64_fix(message_bytes):
    base64_padding = b'=' * (-len(message_bytes) % 4)  
    try:
        decoded_message = base64.b64decode(message_bytes + base64_padding).decode('utf-8')
        print("Успешное декодирование Base64:", decoded_message)
    except Exception as e:
        print("Не удалось декодировать с исправлением Base64:", e)

def decode_lsb_data(lsb_bits):
    message_bytes = bytearray()
    for i in range(0, len(lsb_bits), 8):
        byte_bits = lsb_bits[i:i + 8]
        if len(byte_bits) < 8:
            break
        byte_val = 0
        for bit in byte_bits:
            byte_val = (byte_val << 1) | bit
        if byte_val == 0:  # Метка окончания
            break
        message_bytes.append(byte_val)

    try:
        message = message_bytes.decode('utf-8')
    except UnicodeDecodeError:
        message = message_bytes.decode('ascii', errors='replace')

    try_base64_fix(message_bytes)

    return message

if __name__ == "__main__":
    # Путь к изображению
    image_path = r"C:\Users\21214\Desktop\1.png"

    # Извлечение потока бит из LSB
    lsb_bits = analyze_lsb_blocks(image_path, channel="R", block_size=50, threshold=0.52)

    # Анализ извлеченного потока бит
    print(f"Длина извлеченного потока бит: {len(lsb_bits)}")
    if lsb_bits:
        bit_count_0 = lsb_bits.count(0)
        bit_count_1 = lsb_bits.count(1)
        print(f"Количество бит 0: {bit_count_0}, Количество бит 1: {bit_count_1}")
        print(f"Доля бит 1 в потоке: {bit_count_1 / len(lsb_bits):.4f}")

        # Преобразование потока бит в поток байтов
        message_bytes = bytearray()
        for i in range(0, len(lsb_bits), 8):
            byte_bits = lsb_bits[i:i + 8]
            if len(byte_bits) < 8:
                break
            byte_val = 0
            for bit in byte_bits:
                byte_val = (byte_val << 1) | bit
            if byte_val == 0:  # Метка окончания
                break
            message_bytes.append(byte_val)

        # Сохранение потока бит в файл
        with open("extracted_bits.bin", "wb") as f:
            f.write(message_bytes)
        print("Поток бит сохранен в файл extracted_bits.bin.")

        # Попытка декодирования встроенной информации
        extracted_message = decode_lsb_data(lsb_bits)
        print("Извлеченная информация:", extracted_message)
    else:
        print("Не удалось извлечь поток бит.")
```
